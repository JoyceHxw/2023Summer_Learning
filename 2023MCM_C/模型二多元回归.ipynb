```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from collections import Counter
import math

df = pd.read_csv("D:/2023美赛/2023C/attributes.csv", usecols=['average times', 'repeatimes', 'Syllables', 'word pinlv','word frequency','Morpheme'])
# df=df[:198]
# c=Counter(df["Morpheme"])
df['word frequency']=np.log(df['word frequency'])

Y = df["average times"]
X = df[["repeatimes", "Syllables", "word pinlv",'word frequency']]
X = sm.add_constant(X)
model = sm.OLS(Y, X)
result = model.fit()
result.summary()
```
