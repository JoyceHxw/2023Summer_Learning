## 会员存量、增量分析

```python
month_vip_num=vip_info.pivot_table(index='注册年月',values='会员卡号', aggfunc='count') 
month_vip_num.columns=['月增量']
month_vip_num
```

```python
month_vip_num['存量']=month_vip_num['月增量'].cumsum()
month_vip_num
```

## 增量等级分布

```python
vip_month_level=vip_info.groupby(['注册年月','会员等级']).count()['会员卡号'].unstack()
vip_month_level
```

## 增量等级占比分布

```python
vip_month_level['总计']=vip_month_level.sum(axis=1)
vip_month_level
```

```python
vip_month_level['白银会员占比']=vip_month_level['白银会员']/vip_month_level['总计']*100
vip_month_level['黄金会员占比']=vip_month_level['黄金会员']/vip_month_level['总计']*100
vip_month_level
```

## 地区店均会员数量

```python
store_info=pd.read_excel('./data/门店信息表.XLSX')[['店铺代码','地区编码']]
store_info
```

```python
column_name={'所属店铺编码':'店铺代码'}
vip_info.rename(columns={'所属店铺编码':'店铺代码'},inplace=True)
# vip_info
# store_vip_merge=pd.merge(vip_info,store_info, left_on='所属店铺编码',right_on='店铺代码')
store_vip_merge=pd.merge(vip_info,store_info, on='店铺代码')
store_vip_merge
```

```python
area_vip_num=store_vip_merge[store_vip_merge['地区编码']!='GBL6D01'].groupby('地区编码').count()[['会员卡号']]
area_vip_num.columns=['会员人数']
area_vip_num
```

```python
# 去重
area_vip_num['店铺数']=store_vip_merge[store_vip_merge['地区编码']!='GBL6D01'].drop_duplicates('店铺代码').groupby('地区编码').count()[['店铺代码']]
area_vip_num['每店平均会员数']=round(area_vip_num['会员人数']/area_vip_num['店铺数'],0)
area_vip_num['总平均会员数']=round(area_vip_num['会员人数'].sum()/area_vip_num['店铺数'].sum(),0)
area_vip_num
```

## 各地区会员连带率分析

## 会员复购率分析

```python
area_consume_day=area_consume[['年月','订单日期','卡号','地区编码']]
area_consume_day=area_consume_day.drop_duplicates()
# area_consume_month=area_consume_day.groupby(['年月','地区编码','卡号']).count()
area_consume_month=area_consume_day.pivot_table(index=['年月','地区编码','卡号'],values='卡号',aggfunc='count').reset_index()
area_consume_month.rename(columns={'订单日期':'消费次数'},inplace=True)
area_consume_month['是否复购']=area_consume_month['消费次数']>1
area_consume_month
```

```python
# 按地区统计复购人数和消费人数
reconsume_area_month=area_consume_month.pivot_table(
    values=['消费次数', '是否复购'],
    index='地区编码', aggfunc={'消费次数': 'count', '是否复购': 'sum'})
reconsume_area_month.columns=['复购人数','消费人数']
reconsume_area_month
```

```python
reconsume_rate_area=reconsume_area_month['复购人数']/reconsume_area_month['消费人数']
reconsume_rate_area
```

```python
# 封装成函数
def reconsume_rate(dt1,dt2):
    area_consume_day=area_consume[['年月','订单日期','卡号','地区编码']]
    area_consume_day=area_consume_day[(area_consume_day['订单日期']>=dt1) & (area_consume_day['订单日期']<=dt2)]
    area_consume_day=area_consume_day.drop_duplicates()
    area_consume_month=area_consume_day.pivot_table(index=['年月','地区编码','卡号'],values='卡号',aggfunc='count').reset_index()
    area_consume_month.rename(columns={'订单日期':'消费次数'},inplace=True)
    area_consume_month['是否复购']=area_consume_month['消费次数']>1
    reconsume_area_month=area_consume_month.pivot_table(
        values=['消费次数', '是否复购'],
        index='地区编码', aggfunc={'消费次数': 'count', '是否复购': 'sum'})
    reconsume_area_month.columns=['复购人数','消费人数']
    return reconsume_area_month
```
